```python
import numpy as np
import matplotlib.pyplot as plt

# Set the frequencies for the cosine signals
frequencies = [1, 2, 3]  # Add more frequencies as needed

# Set the time range and interval
t_start = 0
t_end = 2 * np.pi
interval = 0.01

# Generate the time values
t = np.arange(t_start, t_end, interval)

# Generate the cosine signals
signals = []
for freq in frequencies:
    signal = np.cos(2 * np.pi * freq * t)
    signals.append(signal)

# Plot the individual cosine signals
plt.figure()
for i, signal in enumerate(signals):
    plt.plot(t, signal, label='Signal {}'.format(i+1))

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Multiple Cosine Signals')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Calculate the sum of all the signals
sum_signal = np.sum(signals, axis=0)

# Plot the sum of the signals
plt.figure()
plt.plot(t, sum_signal, label='Sum of Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sum of Cosine Signals')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the Fourier transform
fourier = np.fft.fft(sum_signal)
freq = np.fft.fftfreq(len(sum_signal), interval)

# Plot the frequency spectrum
plt.figure()
plt.plot(freq, np.abs(fourier))
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('Frequency Spectrum')
plt.grid(True)
plt.show()
```

```python
# Perform the inverse Fourier transform
inverse_fourier = np.fft.ifft(fourier)

# Plot the original signal
plt.figure()
plt.plot(t, inverse_fourier.real)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Signal')
plt.grid(True)
plt.show()
```
